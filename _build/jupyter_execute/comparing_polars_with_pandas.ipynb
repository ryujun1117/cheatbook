## Reading & writing

### Reading

```python
# pandas
cm_pd = pd.read_csv("customer_master.csv")
cm_pd.head()
```

```python
# polars
cm_pl = pl.read_csv("customer_master.csv")
cm_pl.head()
```

### Writing

```python
# pandas
cm_pd.to_csv("output.csv")
```

```python
# polars
cm_pl.write_csv("output.csv")
```

## Expressions

### select

```python
# pandas
cm_pd[["customer_id", "customer_name"]]
```

```python
# polars
cm_pl.select(pl.col("customer_id", "customer_name"))
```

### Filter

```python
# pandas
cm_pd.loc[(cm_pd["gender"] == "M") & (cm_pd["pref"] == "石川県"), :]
```

```python
cm_pl.filter((pl.col("gender") == "M") & (pl.col("pref") == "石川県"))
```

### add_columns

```python
# pandas
cm_pd["age_avg"] = cm_pd["age"].mean()
cm_pd.head()
```

```python
# polars
cm_pl = cm_pl.with_columns(pl.col('age').mean().alias("age_avg"))
cm_pl.head()
```

### Group by

```python
# pandas
cm_pd[["gender", "age"]].groupby("gender").mean()
```

```python
# polars
cm_pl.group_by("gender", maintain_order=True).agg(
    pl.col("age").count().alias("count"),
    pl.col("age").mean().alias("mean")
)
```

### Combination

```python
# pandas
cm_pd["age_index"] = cm_pd["age"] / cm_pd["age_avg"]
cm_pd.drop(["age", "age_avg"], axis=1)
```

```python
# polars
cm_pl.with_columns((pl.col("age") / pl.col("age_avg")).alias("age_index")).select(
    pl.all().exclude(["age", "age_avg"])
)
```

## Combining DataFrame

### Concat

```python
# pandas
tran1_pd = pd.read_csv("transaction_1.csv")
print(tran1_pd.shape)
tran2_pd = pd.read_csv("transaction_2.csv")
print(tran2_pd.shape)
tran_pd = pd.concat([tran1_pd, tran2_pd], ignore_index=True)
tran_pd.shape
```

```python
# polars
tran1_pl = pl.read_csv("transaction_1.csv")
print(tran1_pl.shape)
tran2_pl = pl.read_csv("transaction_2.csv")
print(tran2_pl.shape)
tran_pl = tran1_pl.vstack(tran2_pl)
tran_pl.shape
```

### Join

```python
# pandas
joined_pd = tran_pd.merge(cm_pd, on="customer_id", how="left")
joined_pd.head()
```

```python
# polars
joined_pl = tran_pl.join(cm_pl, on="customer_id", how="left")
joined_pl.head()
```
